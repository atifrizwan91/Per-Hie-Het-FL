```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics 
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
import time
import math
# from tensorflow import keras
from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score, r2_score
from keras.models import Sequential
from keras.layers import Dense, Dropout,Input
from pathlib import Path
from keras.layers import LSTM
from keras.layers import Activation 
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from keras.layers import Embedding, Flatten, Conv1D, SpatialDropout1D
# from keras.layers import Embedding,Conv1D,LSTM,Input,TimeDistributed,,Flatten,Dropout
from tensorflow.keras.optimizers import Adam
from sklearn import preprocessing
from tensorflow.keras.utils import to_categorical
from keras.metrics import categorical_crossentropy
from sklearn.metrics import f1_score
```

```python
# labels = ['syn_data_0','syn_data_1',"syn_data_2","syn_data_3","syn_data_4", "syn_data_5","syn_data_6","syn_data_7","syn_data_8","syn_data_9"]
labels = ['India', 'Italy', 'Pakistan', 'Philippines', 'Portugal', 'Singapore', 'Sweden', 'Thailand', 'Tunisia','UK',"Scales","Medium_US"]
def get_data(c):
       # df = pd.read_csv("D:\\Projects\Federated Learning\\Datasets\\Fed Dataset Thermal (Transfer Learning Paper)\\"+self.country+".csv")
        df = pd.read_csv('../Data/Correlation Based Data/'+c+'.csv')
        y = df['Thermal sensation']
        drop_list = ['Unnamed: 0','Publication (Citation)','Data contributor','Heating strategy_building level',
                     'Year','Koppen climate classification','Climate','Building type','Database',
                     'City','Country','Outdoor monthly air temperature (Â¡C)','Thermal preference','Season','Cooling startegy_building level',
                     'Air movement preference','Humidity preference','Thermal comfort','PMV','Thermal sensation']
        drop = list(set(df.columns).intersection(drop_list))
        X = df.drop(drop, axis = 1)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
        return X_train, X_test, y_train, y_test
        
```

```python
# path = 'Synthetic Data Results/'
path = 'countries results/'
```

# Decision Tree


```python
#labels = ['India', 'Italy', 'Pakistan', 'Philippines', 'Portugal', 'Singapore', 'Sweden', 'Thailand', 'Tunisia','UK',"Scales","Medium_US"]
# labels = ['data_0','data_1',"data_2","data_3","data_4", "data_5","data_6","data_7","data_8","data_9","Scales","Medium_US"]
# labels = ['syn_data_0','syn_data_1',"syn_data_2","syn_data_3","syn_data_4", "syn_data_5","syn_data_6","syn_data_7","syn_data_8","syn_data_9"]
acc_all = []
acc_all = []
for i in labels:

    X_train, X_test, y_train, y_test = get_data(i)
    clf = DecisionTreeClassifier()
    clf = clf.fit(X_train,y_train)
    y_pred_train = clf.predict(X_train)
    dt_acc_train = accuracy_score(y_train, y_pred_train)
    dt_f1_train = f1_score(y_train, y_pred_train, average='weighted')
    
    y_pred = clf.predict(X_test)
    dt_acc = accuracy_score(y_test, y_pred)
    dt_f1 = f1_score(y_test, y_pred, average='weighted')
    t = [i,dt_acc,dt_f1,dt_acc_train,dt_f1_train]
    acc_all.append(t)
df = pd.DataFrame(acc_all, columns = ['Client','DT Acc','DT F1','DT Acc Train','DT F1 Train'])
df.to_csv(path+'Decision Tree Normal  performance.csv')
df
```

# Decision Tree Entropy

```python

acc_all = []
for i in labels:
    X_train, X_test, y_train, y_test = get_data(i)
    clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)
    clf = clf.fit(X_train,y_train)
    y_pred_train = clf.predict(X_train)
    dt_acc_train = accuracy_score(y_train, y_pred_train)
    dt_f1_train = f1_score(y_train, y_pred_train, average='weighted')
    
    y_pred = clf.predict(X_test)
    dt_acc = accuracy_score(y_test, y_pred)
    dt_f1 = f1_score(y_test, y_pred, average='weighted')
    t = [i,dt_acc,dt_f1,dt_acc_train,dt_f1_train]
    acc_all.append(t)
df = pd.DataFrame(acc_all, columns = ['Client','DT Acc','DT F1','DT Acc Train','DT F1 Train'])
df.to_csv(path+'Decision Tree Entropy  performance.csv')
```

```python
acc_all = []
from sklearn import svm
for i in labels:
    X_train, X_test, y_train, y_test = get_data(i)
    rbf = svm.SVC(kernel='rbf', gamma=0.5, C=0.1).fit(X_train, y_train)
    poly = svm.SVC(kernel='poly', degree=3, C=1).fit(X_train, y_train)
    linear = svm.SVC(kernel='linear', degree=3, C=1).fit(X_train, y_train)
    
    poly_pred_train = poly.predict(X_train)
    rbf_pred_train = rbf.predict(X_train)
    linear_pref_train = linear.predict(X_train)
    
    poly_accuracy_train = accuracy_score(y_train, poly_pred_train)
    poly_f1_train = f1_score(y_train, poly_pred_train, average='weighted')
    rbf_accuracy_train = accuracy_score(y_train, rbf_pred_train)
    rbf_f1_train = f1_score(y_train, rbf_pred_train, average='weighted')
    linear_accuracy_train = accuracy_score(y_train, linear_pref_train)
    linear_f1_train = f1_score(y_train, linear_pref_train, average='weighted')
    
    poly_pred = poly.predict(X_test)
    rbf_pred = rbf.predict(X_test)
    linear_pref = linear.predict(X_test)
    poly_accuracy = accuracy_score(y_test, poly_pred)
    poly_f1 = f1_score(y_test, poly_pred, average='weighted')
    rbf_accuracy = accuracy_score(y_test, rbf_pred)
    rbf_f1 = f1_score(y_test, rbf_pred, average='weighted')
    linear_accuracy = accuracy_score(y_test, linear_pref)
    linear_f1 = f1_score(y_test, linear_pref, average='weighted')
    
    t = [i,linear_accuracy,linear_f1,rbf_accuracy,rbf_f1,poly_accuracy,poly_f1,
        linear_accuracy_train,linear_f1_train,rbf_accuracy_train,rbf_f1_train,poly_accuracy_train,poly_f1_train]
    acc_all.append(t)
df = pd.DataFrame(acc_all, columns = ['Client','Linear Acc','Linear F1', 'RBF Acc','RBF F1','Poly Acc','Poly F1','Linear Acc Train','Linear F1 Train', 'RBF Acc Train','RBF F1 Train','Poly Acc Train','Poly F1 Train' ])
df.to_csv(path+'SVM performance.csv')
df
```
